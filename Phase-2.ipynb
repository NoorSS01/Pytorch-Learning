## Autograd and Optimization


```python
import torch
```

Automatic differentiation


```python
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2 + 2 * x + 1
y.backward()

print(f"y value: {y.item()}")
print(f"dy/dx at x=3: {x.grad.item()}")
```

Gradient descent on a simple function


```python
w = torch.tensor(5.0, requires_grad=True)
learning_rate = 0.1

for epoch in range(20):
    loss = (w - 1.5) ** 2
    loss.backward()
    
    with torch.no_grad():
        w -= learning_rate * w.grad
    
    w.grad.zero_()

print(f"Optimized w: {w.item():.4f}")
```

Use torch.optim


```python
param = torch.tensor([10.0], requires_grad=True)
optimizer = torch.optim.SGD([param], lr=0.2)

for step in range(25):
    optimizer.zero_grad()
    loss = (param - 3.0).pow(2).sum()
    loss.backward()
    optimizer.step()

print(f"Parameter after optimization: {param.item():.4f}")
```
