## Saving and Loading PyTorch Models


```python
import torch
import torch.nn as nn
```

Train a small model


```python
torch.manual_seed(42)

X = torch.randn(600, 3)
y = ((1.2 * X[:, 0] - 0.7 * X[:, 1] + 0.4 * X[:, 2]) > 0).long()

model = nn.Sequential(
    nn.Linear(3, 12),
    nn.ReLU(),
    nn.Linear(12, 2)
)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(30):
    logits = model(X)
    loss = criterion(logits, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

print("Model training complete")
```

Save and load state_dict


```python
torch.save(model.state_dict(), 'pytorch_model_state.pth')

loaded_model = nn.Sequential(
    nn.Linear(3, 12),
    nn.ReLU(),
    nn.Linear(12, 2)
)
loaded_model.load_state_dict(torch.load('pytorch_model_state.pth'))
loaded_model.eval()

print("Model state loaded successfully")
```

Compare predictions


```python
sample = torch.tensor([
    [0.8, -0.1, 0.2],
    [-0.2, 0.6, -0.5],
    [1.1, -0.9, 0.3]
], dtype=torch.float32)

with torch.no_grad():
    pred_1 = model(sample).argmax(dim=1)
    pred_2 = loaded_model(sample).argmax(dim=1)

print("Original model predictions:", pred_1.tolist())
print("Loaded model predictions:", pred_2.tolist())
```
