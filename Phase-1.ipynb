## PyTorch Basics and Tensors


```python
import torch
import numpy as np
```

Create tensors


```python
scalar = torch.tensor(10)
vector = torch.tensor([1, 2, 3, 4])
matrix = torch.tensor([[1, 2], [3, 4]])

print(f"Scalar: {scalar}")
print(f"Vector: {vector}")
print(f"Matrix:{matrix}")
```

### Tensor properties


```python
print(f"Shape: {matrix.shape}")
print(f"Dimension: {matrix.ndim}")
print(f"Data type: {matrix.dtype}")
print(f"Number of elements: {matrix.numel()}")
```

### Tensor operations


```python
a = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
b = torch.tensor([[5.0, 6.0], [7.0, 8.0]])

print(f"Addition:{a + b}")
print(f"Element-wise multiplication:{a * b}")
print(f"Matrix multiplication:{a @ b}")
print(f"Mean of a: {a.mean()}")
```

### Reshape, transpose, and type conversion


```python
x = torch.arange(12)
reshaped = x.reshape(3, 4)
transposed = reshaped.T
float_tensor = reshaped.float()

print(f"Original tensor:{x}")
print(f"Reshaped:{reshaped}")
print(f"Transposed:{transposed}")
print(f"Tensor dtype after conversion: {float_tensor.dtype}")
```

### Random tensors and seed


```python
torch.manual_seed(42)
random_tensor = torch.randn(2, 3)
print(random_tensor)
```
